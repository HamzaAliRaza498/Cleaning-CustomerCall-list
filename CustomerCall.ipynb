```python
import numpy as np
import pandas as pd
load = pd.read_excel("Customer Call List.xlsx")
```

```python
load
```

```python
#FIRST REMOVE DUPLICATE USING DUPLICATE METHOD
load = load.drop_duplicates()
load
```

```python
load = load.drop(columns="Not_Useful_Column") #Remove unnessary calumn
load
```

```python
load["Last_Name"] = load["Last_Name"].str.strip("123._/") #Remove all the unnessary things like dots(.),underscore(_),slash(/) from Last name 
load 
```

```python
load["Phone_Number"] = load["Phone_Number"].str.replace('[^a-zA-Z0-9]', '',regex=True )
load
```

```python
def format_phone_number(x):
    if isinstance(x, str) and len(x) == 10:
        return f"{x[0:3]}-{x[3:6]}-{x[6:10]}"
    return x

load["Phone_Number"] = load["Phone_Number"].apply(format_phone_number)
load
```

```python
load["Phone_Number"] = load["Phone_Number"].str.replace('Na', '', regex=False)
load["Phone_Number"] = load["Phone_Number"].str.replace('nan', '', regex=False)
load
```

```python
load
```

```python
# Ensure all values are strings
load["Address"] = load["Address"].astype(str)

# Split the 'Address' column by the first comma and expand into two columns
split_columns = load["Address"].str.split(',', n=1, expand=True)

# Assign new columns to the DataFrame
load[['Streets', 'Post code']] = split_columns
load
```

```python
# Drop the 'Address' column from the DataFrame
load = load.drop(columns=['Address'])
load
```

```python
# Replace occurrences of the string 'NaN' with an empty string in the 'Last_name' column
load['Last_Name'] = load['Last_Name'].str.replace('NaN', '', regex=False)
load
```

```python
load
```

```python
load['Phone_Number'] = load['Phone_Number'].str.replace('NaN', '', regex=False)
load
```

```python

# Replace 'NaN' in the Phone_Number column with an empty string
load['Phone_Number'] = load['Phone_Number'].replace('NaN', '', regex=False)

# Remove any remaining NaNs in the Phone_Number column
load['Phone_Number'] = load['Phone_Number'].fillna('')
load
```

```python
# Replace 'NaN' in the Last_Name column with an empty string
load['Last_Name'] = load['Last_Name'].replace('NaN', '', regex=False)

# Remove any remaining NaNs in the Last_Name column
load['Last_Name'] = load['Last_Name'].fillna('')

# Print the DataFrame to verify changes
load
```

```python
load["Paying Customer"]  =load["Paying Customer"].str.replace('Yes','Y')
load["Paying Customer"] =load["Paying Customer"].str.replace('No','N')
load
```

```python
load["Do_Not_Contact"]  =load["Do_Not_Contact"].str.replace('Yes','Y')
load["Do_Not_Contact"]  =load["Do_Not_Contact"].str.replace('No','N')
load
```

```python
# Replace 'NaN' in the Do_Not_Contact column with an empty string
load["Do_Not_Contact"] = load["Do_Not_Contact"].str.replace('NaN', '', regex=False)

# Remove any remaining NaNs in the Do_Not_Contact column
load["Do_Not_Contact"] = load['Do_Not_Contact'].fillna('')

# Print the DataFrame to verify changes
load
```

```python
# Replace 'NaN' in the Do_Not_Contact column with an empty string
load["Paying Customer"] = load["Paying Customer"].str.replace('N/a', 'N', regex=False)

# Remove any remaining NaNs in the Do_Not_Contact column
load["Paying Customer"] = load['Paying Customer'].fillna('N')

# Print the DataFrame to verify changes
load
```

```python
# Iterate over the indices of the DataFrame
for x in load.index:
    # Check if the value in 'Do_Not_Contact' for the current row is 'Y'
    if load.loc[x, 'Do_Not_Contact'] == 'Y':
        # Drop the row where 'Do_Not_Contact' is 'Y'
        load.drop(index=x, inplace=True)

# Print the DataFrame to verify changes
load
```
